```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd

%matplotlib inline
```

```python
digits = load_digits()
```

```python
digits.data.shape
```

# Showing the Images and Labels

```python
plt.figure(figsize=(20,4))
for index,(image,label) in enumerate(zip(digits.data[0:5],digits.target[0:5])):
    plt.subplot(1,5,index+1)
    plt.imshow(np.reshape(image,(8,8)),cmap='Blues')
```

#  Splotting data into training set and test sets

```python
# test_size: What proportion of original data is used for test set
x_train,x_test,y_train,y_test=train_test_split(
    digits.data,digits.target,test_size=0.25,random_state=0
)
```

```python
print(x_train.shape)
```

```python
print(y_train.shape)
```

```python
print(x_test.shape)
```

```python
print(y_test.shape)
```

# Scikit_learn 4 Step Modeling Pattern

Step 1: Import the model you want to use

```python
from sklearn.linear_model import LogisticRegression
```

Step 2: Make the instance of the Model

```python
logisticRegr = LogisticRegression()
```

```python
logisticRegr
```

Step 3: Training the model on the data ,storing the information learned from the data
Model is learning the relationship between x (digit) and y (labels)

```python
logisticRegr.fit(x_train,y_train)
```

Step 4: Predict the labels of new data (new images)

Uses the informationthe model learned during the model training process

```python
# Returns a Numpy Array
# Predict for one observation(image)
logisticRegr.predict(x_test[0].reshape(1,-1))
```

```python
# Predict for Multiple Observaation(image)
logisticRegr.predict(x_test[0:10])
```

```python
# Make Prediction on entire test data
predictions = logisticRegr.predict(x_test)
```

```python
predictions.shape
```

# Measuring Model Performance

accuracy(fracton of correct predictions): correct predictions/totalnumber of data points

Masically how the modelperforms on new data(test set)

```python
# use score method to get accuracy of model
score = logisticRegr.score(x_test,y_test)
print(score)
```
